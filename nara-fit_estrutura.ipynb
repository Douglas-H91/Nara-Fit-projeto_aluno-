## Estrutura básica de cadastramento
 1 - Definindo dicionários: dados dos alunos cadastrados

```python
alunos_cadastrados = {
    "nome": [],
    "idade": [],
    "peso": [],
    "Altura": [],
    "imc": [],
    "objetivo": []
}
```

2 - Definindo as funções. Aplicação de Tratamento de Exceções, Looping While True e Return.

```python

def saudacao ():
    print('Bem-vindo à Academia Nara Fit! Escolha a opção desejada: ')

def calculo_imc(peso, altura):
    imc = peso / (altura * altura)
    return imc

def cadastrar_aluno():
    nome = input('Digite o nome do aluno que deseja cadastrar: ').title()
    while True:
        try:
            idade = int(input('Digite a idade do aluno: '))
            break
        except ValueError:
            print('Digite apenas numero inteiros')
        return idade

def cadastrar_peso():
    while True:
        try:
            peso = float(input('Digite o peso do aluno (kg): '))
            break
        except ValueError:
            print('Digite apenas numero inteiros')
    return peso

def cadastrar_altura():
    while True:
        try:
            altura = float(input('Digite a altura do aluno (m): '))
            if altura > 2.10:
                print('Altura inválida. Digite apenas valores em metros (ex: 1.69)')
            else:
                break
        except ValueError:
            print('Digite apenas números')
    return altura

def cadastrar_objetivo():
    while True:
        try:
            objetivo = input('Digite o objetivo do aluno: ').title()
            if objetivo not in ['Emagrecimento', 'Hipertrofia', 'Condicionamento Físico']:
                print('Objetivo inválido. Digite apenas Emagrecimento, Hipertrofia ou Condicionamento Físico')
            else:
                break
        except ValueError:
            print('Digite apenas letras')
        print(f"Aluno {nome} cadastrado com sucesso!")
        return objetivo

def listar_alunos():
    if len(alunos_cadastrados) == 0:
        print('Nenhum aluno cadastrado') 
    else:
        for i in alunos_cadastrados:
            print(f'Aluno {i}')
```

3 - Estrutura do Looping While True

```python
saudacao ()
while True:
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Sair")
    opcao = int(input("Escolha uma opcao: "))

    match opcao:
        case 1:
            cadastrar_aluno()
        case 2:
            listar_alunos()
        case 3: 
            break

        
```
